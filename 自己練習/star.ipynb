```python
# 題目 4（對稱金字塔）-2
#     *
#    ***
#   *****
#  *******
# *********

n = 5
for x in range(n):
    for y in range(2*n-1):
        if x+y >= n-1 and x-y >= -n+1:
            print("*",end="")
        else:
            print(" ",end="")
    print()
```

```python
# 題目 5 (半邊沙漏) -2
# *****
#  ****
#   ***
#    **
#     *
#    **
#   ***
#  ****
# *****
n = 9 
a = (n-1)//2
for x in range(n): 
    for y in range(n-a): 
        if x - y <= 0 or x + y >= n-1: 
            print("*",end="") 
        else: 
            print(" ",end="") 
    print()
```

```python
# 題目 8 (正方形邊界)
# *****
# *   *
# *   *
# *   *
# *****

n = 5
for x in range(n):
    for y in range(n):
        if x == 0 or x == n-1 or y == 0 or y == n-1:
            print("*" , end="")
        else:
            print(" ", end="")
    print()
```

```python
# 題目 9 (正方形邊框)
# *****
# *   *
# *   *
# *   *
# *****

n = 5
for x in range(n):
    for y in range(n):
        if (abs(x-2) + abs(y-2)) >= 3:
            print("*",end="")
        else:
            print(" ", end="")
    print()
```

```python
# 題目 10 (圓形)


n = 30
a = n//2
for x in range(n):
    for y in range(n):
        if (x-a)**2 + (y-a)**2 <= (a)**2:
            print("*", end="")
        else:
            print(" ", end="")
    print()


```
