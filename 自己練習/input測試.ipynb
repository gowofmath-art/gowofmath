```python
# 測試 input 測試質數

n = int(input("請輸入一個整數: "))

prime = True

for i in range(2, n):
    if n % i == 0:
        prime = False
        break

if prime and n > 1:
    print("是質數")
else:
    print("不是質數")
```

```python
n = int(input("請輸入一個整數: "))

for x in range(2,n):
    for y in range(2,x):
        if x % y == 0:
            break
    else:
        print(x,end=",")
```
