```python
n = int(input(">"))

for i in range(n):
    for j in range(n):
        if i==j or i+j == n-1:
            print('x',end="")
        else:
            print(' ',end="")
    print()
```

```python
n = int(input(">"))

for x in range(n):
    for y in range(2*n-1):
        if y >= x and x + y <= 2*n-2 :
            print(x,end="")
        else:
            print(' ',end="")
    print()
```

```python
# 不在99乘法表裡面的2位數
i=0
for n in range(10,100): # 兩位數迴圏

    found = True #先設定found「找到」為真
    for x in range(2,10): #除數範圍x由2到9
        if n%x == 0 and n//x < 10: #當x能整除n且商為個位數
            found = False # 排除數字Il'設定found為假
            break #並提早跳離迴圏

    if found: #當found「找到」仍為真
        i += 1
        print(n , end=" ")
        if i%20 == 0 : print() #每20個數換列
print()
```

```python
# 二位數中 扣除 3,5,7的倍數
i = 1
for n in range(10,100):
    if n%3 == 0 or n%5 == 0 or n%7 == 0 : continue
    print(i , n)
    i += 1
```

```python
# 二位數中 扣除 3,5,7的倍數 也可以這樣寫！

i = 1
for n in range(10,100):
    if not (n%3 == 0 or n%5 == 0 or n%7 == 0) : 
        print(i , n)
        i += 1
```
