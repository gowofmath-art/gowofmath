```python
# 找質數

n = 100
for x in range(2,n):
    for y in range(2,x):
        if x % y == 0:
            break
    else:
        print(x,end=",")
```

```python
# 找質數

primes = []  # 準備一個空清單來放質數

for n in range(2, 101):  # 從 2 跑到 100
    is_prime = True      # 先假設它是質數
    for i in range(2, n): # 拿 2 到 n-1 的數字來除除看
        if n % i == 0:    # 如果餘數為 0，代表除得盡
            is_prime = False  # 它不是質數
            break         # 既然找到了因數，就不用再往下測了，跳出內層迴圈
    if is_prime:         # 如果跑完檢查，它依然是 True
        primes.append(n)  # 把這個數加進清單

print(primes)
```

```python
# 找質數

primes = []
for n in range(2, 101):
    for i in range(2, int(n**0.5) + 1): # 只檢查到平方根
        if n % i == 0:
            break
    else:
        primes.append(n)
print(primes,end=",")
```

```python
# 找質數 一行結束

prime = {n for n in range(2,101) if all(n%i !=0 for i in range(2, int(n**0.5)+1))}
print(prime)
```

```python
# 找合數

n = 100
for x in range(2,n):
    for i in range(2, int(x**0.5)+1):
        if x % i == 0:
            print(x,end=",")
            break

```

```python
# 找合數 一行結束

composite = {n for n in range(2,101) if any(n%i ==0 for i in range(2,int(n**0.5)+1))} 
print(composite)
```
