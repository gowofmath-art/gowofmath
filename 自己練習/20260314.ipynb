```python
# 找出三位數的數字和為10且數字都不同的所有三位數，例如：325、910，驗證共有40個數
def find_three_digit_numbers():
    count = 0
    result = []
    for i in range(1, 10):  # 百位數不能為 0
        for j in range(10):
            for k in range(10):
                if i != j and i != k and j != k and i + j + k == 10:
                    count += 1
                    result.append(i * 100 + j * 10 + k)
    return count, result

count, numbers = find_three_digit_numbers()
print(f"各位數字之和為 10 且各位數字都不同的三位數共有 {count} 個。")
print(f"這些數字包括：{numbers}")

if count == 40:
    print("驗證結果：共有 40 個，與題目描述相符。")
else:
    print("驗證結果：共有", count, "個，與題目描述不符。")
```
