```python
# 類別名稱
class Car:
    wheels = 4 # 類別屬性 (Class Attribute)
    # 建構函式 > magic method
    def __init__(self,brand,model,door=4):
        self.brand = brand 
        self.model = model
        self.door = door


# 實例屬性

# 方法
    def car_info(self):
        print(f"info:Brand{self.brand} Model:{self.model} has:{self.door} doors")

print(Car.wheels) 
# 使用car的類別屬性

# 建立實例 參數要設定好
car1 = Car("benz","c300")
car2 = Car("toyota","altis",5)
print(car1.model)
print(car2.model)
print(car1.door)

car1.car_info()
car2.car_info()
```
