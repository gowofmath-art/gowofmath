### lambda 匿名函式  
用法  

`<function name> = lambda <arguments>: <expression>

````python
# 建立函式
def add(x,y):
    return x+y

a = add(139,103)
print(a)
````

````python
# 這樣才對
def add(x,y):
    return x+y

print(add(139,103))
````

````python
# 基本用法
add = lambda x,y: x+y
a = add(131,69)
print(a)
````

````python
#  return index 2 的data
def get_index_2(data):
    return data[2]
get_index_2 = lambda data:data[2]

result = get_index_2([10,20,30,40])
print(result)
````

````python
# 條件判斷
def abs(n):
    if n >= 0:
        return n
    else:
        return -n
    
result = abs(-100)
print(result)
````

### 小試身手  
請建立一個 list ，並在該 list 內產生多個 `[1, 2, x]` 的 list (x 的值要遞增)  
x 到 10 即可  
Ex :  
```Python
[ [1, 2, 1], [1, 2, 2], [1, 2, 3]...[1, 2, 10] ]
```

````python
list_2=[]
for x in range(1,11):
    list_1 = [1,2,x]
    list_2.append(list_1)

print(list_2)
# 重點是要把每一次 的 list_1 放進去 list_2
````

````python
# 令解

A = [[1,2,i] for i in range(1,11)]
print(A)
````

##### `sort()` 方法

````python
n = [10,5,59,3,100]
n.sort()

print(n)
print(n[0])  # 最小的意思
print(n[-1]) # 最大的意思
````

````python
result = [[1,2,i] for i in range(1,11)]
print(result)
result.sort(reverse=True)   # 從大到小
print(result)
````

每三個 list 中間的值要改成 5  
Ex:  
```python
[ [1, 2, 1], [1, 2, 2], [1, 5, 3], [1, 2, 4], [1, 2, 5], [1, 5, 6]...]
```

````python
list_2=[]
for x in range(1,11):
    list_1 = [1,2,x]
    # list_1[1] = 5
    list_2.append(list_1)

print(list_2)
# list_2[list_1[0]]=[2,3,5]
# list_2[list_1[1]]=[1,]
# list_2[0][2]=99
for y in range(2,11,3):
    list_2[y][1]=5
print(list_2)

# list_2[list_1[0]]=[2,3,5] 原因是因為 list_1[0] = 1 變成 修改 list_2[1] 
# list_2[0][2]=99 這樣才可以針對 第一個 [1,2,1] 改為 [1,2,99]
````

````python
list_2=[]
for x in range(1,11):
    list_1 = [1,2,x]
    list_2.append(list_1)
print(list_2)
for y in range(2,11,3):
    list_2[y][1]=5
print(list_2)
````

````python
# 老師解法
result = []
for i in range(1,11):
    if i % 3 == 0:
        result.append([1,5,i])
    else:
        result.append([1,2,i])

print(result)
````
