```python
# 建立 function 3 有參數
def greenting(name, greet):
    print(f"neko {name},{greet}")

greenting("pochita", "hello")
greenting("Denji","good")
```

```python
# 建立 function 4 預設參數值
def greenting(name, greet= "Good night"):
    print(f"neko {name},{greet}")

greenting("")
greenting("apple","sega")
# 沒有預設參數值，再使用時就一定要給該參數
```
