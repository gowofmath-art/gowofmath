```python
# 全域變數
name = "naruto"

# 定義函式
def greeting():
    name = "pochita"
    print(f"hello,{name}")

# 執行函式
greeting()

# 輸出 name
print(name)

# 看到了嗎? 就算我在一開始將 name 令為 naruto ，也不影響我在def 那邊的name
```

##### 使用 global 讓函式存取全域變數的值

```python
# 全域變數
name = "naruto"

# 定義函式
def greeting():
    global name 
    name = "pochita"
    print(f"hello,{name}")

# 執行函式
greeting()

# 輸出 name
print(name)

# 使用global 可修改全部的變數 都變成 "pochita"
```
