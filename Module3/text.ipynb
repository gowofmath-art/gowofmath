```python
# 練習題：神祕的 7 號收藏家
# 請嘗試用 while 迴圈 寫出以下邏輯：
# 情境：從 1 開始往後找數字。

# 累加條件：只累加 包含數字 7 的奇數（為了簡化，我們先練習：尾數是 7 的奇數，例如 7, 17, 27...）。
# 停止條件：當你累加的總和 (total) 超過 500 時，就停止迴圈。
# 最後結果：印出最終的 總和 以及 最後一個被加進去的數字。

x = 0
i = 1
while x <=500:
    if i % 10 == 7:
        x = x + i
    i = i + 1
print(x)
print(i-1)        


```
