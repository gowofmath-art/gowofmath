條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

```python
fruit = "apple"

if fruit == "apple":
   print("fruit is apple")
```

##### if...else...

```python
fruit = "banana"

if fruit == "apple":
   print("fruit is apple")
else:print("fruit is not apple")
```

##### if...elif... (else)

```python
fruit = "banana"

#當前面已經有符合 後面就不判斷
if fruit == "apple":
   print("fruit is apple")
elif fruit == "banana": #執行另一個程式碼
   print("fruit is banana")
elif fruit != "apple":
   print("fruit is not apple")
else:print("fruit is not apple") #當if語句沒有達成 ，就執行 else 這段
```
