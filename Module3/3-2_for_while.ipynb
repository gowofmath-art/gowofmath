| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python
data = "contents"

for apple in "contents":
    if apple == "t":
       break #立即結束當前的迴圈 不論剩幾次
    print(apple,end=",")
```

```python
data = "contents"

for apple in data:
    if apple == "t":
       continue #立即結束當前的迴圈 不論剩幾次
    print(apple,end=",")
```

```python
s = [1, 2, 3]
for i in s:
    print(i,end=",")
    if i = 2:
    break #立即結束當前的迴圈 不論剩幾次
```

```python
for i in range(5):
    print("hello")
```

```python
# range(m) 
# 走訪 0-m-1 的數字

for i in range(5):
    print(i+1,end=",")
```

```python
# range(n, m)

for i in range(6,10):
    print(i,end=",")
```

```python
# range(n, m-1, step)

for i in range(4,100,4):
    print(i,end=",")
```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

```python
a = 10 
n = 10
if a%2 == 0:
    for i in range(a, a+2*n,2):
        print(i,end=",")
elif a%2 == 1:
    for i in range(a+1, a+1+2*n,2):
        print(i,end=",")  
```

##### while

```python
x = 1
while x <= 5:
    print(x,end="")
    x = x+1
```

```python
x = 1
while x <= 5:
    x = x+1
    print(x,end="")
```

```python
x = 1
while True:  #不管怎樣就是會一直執行底下的程式
    if x > 5:
        break
    print(x)
    x = x+1
```
