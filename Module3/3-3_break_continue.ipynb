##### break

```python
'''
當偵測到字母 t 時，就會強制結束迴圈
'''

```

```python
data = "contents"

for apple in data:
    if apple == "t":
       break #立即結束當前的迴圈 不論剩幾次
    print(apple,end=",")
```

##### continue

```python
'''
當偵測到字母 t 時，
會跳過本次迴圈剩下的程式碼 print(char)，
但不會結束迴圈，仍然會進入下一圈繼續執行
'''
data = "contents"

for apple in data:
    if apple == "t":
       continue #立即結束當前的迴圈 不論剩幾次
    print(apple,end=",")
```
