##### try...except...

```python
# try except 不要濫用，除非確定這個錯誤可以被忽略的
# 有錯就該處理，而不是一直用try except
try:
    x = 120
    print(y)
    print(x)
except:
    print("程式發生錯誤")
# 當try底下發生錯誤，就要執行程式碼
```

```python
# 可以連續定義數個例外處理機制
# 最符合的例外事件類別會先拋出
try:
    x = 120
    print(y)
    print(x)
except NameError as e:
    print(f"發生 NameError:{e}")
except Exception as e:
    print(f"發生 IndexError:{e}")

except Exception as e:
    print(f"程式發生錯誤:{e}")

# 告訴我們哪裡有錯
# 只會發生第一個except

print("繼續執行啊")
print("戰鬥續行！")
# 要繼續執行也可以，不用整個都在try 迴圈中
```

##### try...except...else...finally

```python
# 如果沒有錯誤，就會執行 else 程式區塊
# finally 程式區塊是無論有無錯誤，都會執行

try:
    x = 100
    y = [1,2]
    print(y[2])
    print(x)

except Exception as e:
    print(f"發生 Exception:{e}")
else:
    print("沒有錯誤 才會執行")
finally:
    print("無論有無錯誤，都會執行")
```
