### `json` 套件

| 用法 | 說明 |
|:---:|:---:|
|json.dump(<obj>,<file>) |物件轉JSON並寫入文件|
|json.dump(<obj>) |物件轉JSON字串|
|json.load(f) |讀取JSON文件|
|json.loads(f) |將JSON字串轉成Python物件|

```python
import json
n = [i for i in range(1,51) if i %2 ==0]
print(n)

with open('result.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(n))
```

```python
import json
n = [i for i in range(1,51) if i %2 ==0]
print(n)

with open('result.json', 'w', encoding='utf-8') as f:
    json.dumps(n,f)
```

```python
import json

n = [i for i in range(1, 51) if i % 2 == 0]
json_string = json.dumps(n) # 先轉成字串

with open('result.json', 'w', encoding='utf-8') as f:
    f.write(json_string) # 用檔案的 write 方法寫入
print(n)
```

```python
with open('result.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

print(data)
print(type(data))
print(data['0'])
```

```python
with open('result.json', 'r', encoding='utf-8') as f:
    str_data = f.read()

    print(str_data)
    print(type(str_data))
    
# 來源不一定是從檔案，很常需要這樣作轉換

data = json.loads(str_data)
print(type(data))
print(data)
print(data[0])
```

```python
import json
nums = "[1,2,3]"
print(nums)

data = json.loads(nums)
print(data)
print(data[0])
```
