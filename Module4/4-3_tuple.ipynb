```python
# tuple 初始化: 第一種
class_1 = ("AI", "BI")
print(class_1)
```

```python
# tuple 初始化: 第二種
class_1 = "AI", "BI"
print(class_1)
```

```python
a,b = 1,2
print(a)
print(b)
```

```python
# 透過指定索引輸出值
class_1 = "AI", "BI"
print(class_1[1])
```

```python
# 複數變數修改值
a,b,c = 1,2,3
print(a,b,c)

```

```python
# tuple 不可以修改指定索引的值
class_1 = "AI", "BI"
class_1[0] = "CI"
print(class_1[1])
```

```python
# ( )可以修改指定索引的值
class_1 = "AI", "BI"
data = list(class_1)
data[0] = "CI"
print(data)
```

```python
# 用 for 迴圈逐一輸出資料
class_1 = "AI", "BI"
for i in class_1:
    print(i)
```

```python
# 用 len 計算 tuple 資料個數
class_1 = "AI", "BI"

print(len(class_1))
```
