```python
# 字串變數初始化
m = "aple"
print(m)
```

```python
# 替換字串
#因為字串不可換，改字串必須做重新賦值的動作，或是用新變數作賦值
name = "aple"
'''

將 string 中的 str1 替換成 str2
'''
# le換成 ps
name2 = name.replace("le","ps") #會回傳新的資料，所以用變數去接
print(name)
print(name2)
```

```python
# 去除兩側空格
data = "_Hello!World_"
'''

去除字串 string 左、右兩邊的字串
'''
data2 = data.strip("_")
print(data)
print(data2)
# 回傳給我新的資料 (原本的還在)

#只想要去除左邊
data3 = data.lstrip("_")
print(data3)
data4 =data.rstrip("_")
print(data4)
```

```python
# 字串小寫變大寫
msg = "HELLO WORLD"
'''

將字串 string 裡的字母全部改成小寫
'''

lower_msg = msg.lower()
print(lower_msg)

msg = "hello world"
'''
str.upper()
將字串 string 裡的字母全部改成大寫
'''
upper_msg = msg.upper()
print(upper_msg)

```

```python
# 字串分割成 list
greeting = "hello world! hello John"
word_list = greeting.split()
print(word_list)


```

```python
greeting = "hello world! hello John"
data = greeting.replace(" ","_")
print(data)

word_list = data.split("_") #指定作分割
print(word_list)
```
