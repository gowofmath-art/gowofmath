```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['greem', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange']
    }
}
print(fruit_dict)
# 錯誤的寫法  格式不一樣會有問題

# 判斷 某個key是否在dict
new_fruit = "banana"
if "orange" not in fruit_dict:
    print(f"{new_fruit} 不在dict中") 
```

```python
# 另一種方式 dict.get 避免搜尋不到 沒有
print(fruit_dict["apple"])
print(fruit_dict.get("apple"))
print(fruit_dict.get("banana"))
```

```python
# 使用 for 迴圈讀取 dict 資料1 
# 拿出所有的 key (主詞條)
for key in fruit_dict:
    print(key)
```

```python
# 使用 for 迴圈讀取 dict 資料2
# 拿出所有的value (個別的子情況)

for value in fruit_dict.values():
    print(value)

```

```python
# 使用 for 迴圈讀取 dict 資料3  .items
# 同時拿出 key,value

for key,value in fruit_dict.items():
    print(key)
    print(value)

# key 在前 ，value 在後
# 變成 tuple
```

```python
for x,y in fruit_dict.items():
    print(x)
    print(y)
# 這樣也可
```

```python
 # 取得 dict 的 value 拿出所有的color 


for key,value in fruit_dict.items():
    # print(value["color"])
    for color in value["color"]:
        print(color)

# 這樣才可以
```

```python
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]
```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}


fruit_dict = { }

for x in fruit_color_list:
    # print(x)
    for key , value in x.items():
        print(key , value)
        fruit_dict[key] = [value]
        
print(fruit_dict)
#    這樣資料會被替換
    

```

```python
# 另一種做法 .get 的作法

A = {}

for x in fruit_color_list:
    print(x)
    for a ,b in x.items():
        print(a,b)
        if A.get(a) is None:
            A[a] = [b]
        else:
            A[a].append(b)

print(A)

# 輸出一筆看看就好 打break
```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}


fruit_dict = { }

for x in fruit_color_list:
    print(x)
    for a , b in x.items():
        print(a , b)
        if a not in fruit_dict:        
            fruit_dict[a] = [b]
        else:
            fruit_dict[a].append(b)
        # 如果沒建立過這個key
print(fruit_dict)
```

##### `defaultdict(<type>)`

```python
from collections import defaultdict

fruit_dict = defaultdict(list)

print(fruit_dict)

# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
print(fruit_dict['apple'])

# 以int型態的預設值 0 來初始化
print(fruit_dict)

```

```python
from collections import defaultdict
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]

x = defaultdict(list)

for a in fruit_color_list:
    # print(a)
    for i,j in a.items():
        x[i].append(j)

print(x)

x = dict(x)
print(x)
```
