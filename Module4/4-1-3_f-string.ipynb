```python
# f-string (又稱作 formatted string literals, version >= 3.6)
name = "julia"
age = 21

msg = f"{name} is {age} year old."
print (msg)
```

```python
# 靠左對齊、靠右對齊

msg = f"name:[{name:10}],[{age:>10}]"

print(msg)
```

```python
# 靠左對齊、靠右對齊，字數不足，補「ㄏ」
# (沒寫補什麼，預設空格)
msg = f"name:[{name:ㄏ<10}],[{age:ㄏ>10}]" #靠左靠有幾個數字  從第一個字數起來

print(msg)
```

```python
# 置中對齊
msg = f"name:[{name:^10}],[{age:^10}]"

print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
msg = f"name:[{name:<10.2}]]"

print(msg)


```

```python
x = 3.1415926
print(f"{x:.4f}")
```

```python
x = 31415926
print(f"{x:,}")
```
