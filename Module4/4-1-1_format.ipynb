```python
# 多組文字
msg = '%s, %s!' % ('Hello', 'World')
print(msg)
```

```python
# 整數
msg = 'I am %d years old.' %5
print(msg)
```

```python
# 文字與整數
msg = '%s is %d years old.' %('andy',18)
print(msg)
```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊)
msg = '%10s' %'hello'
print(msg)
```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)
msg = '%-10s' % 'Hello'
print(msg)
```

```python
# 指定浮點數位數
msg = '%.3f' % 12.3456
print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
msg = '%.2s' % 'Hello'
print(msg)
```

```python
# 空白補 0
msg = '%06.2f' % 3.1415926
print(msg)
```
