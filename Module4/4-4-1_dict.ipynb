```python
# 建立一個 value 含有 dict 的 dict
dict2 = {
    "name" : "ako",
    "age" : 28,
    "love_color" : "black" ,
    "info" : {
        "phone_number" : ["0931415926", "092718128"],
        "nick_name" : "big"
    }
}
print(dict2)

# 可以想像 集合中還有集合！
```

```python
# 取得 nickname
print(dict2["info"])
print(f"info:{dict2["info"]}")
print(f"info:{dict2["info"]["nick_name"]}")
```

```python
# 取得所有 phone_number
print(dict2["info"]["phone_number"])
for phone in dict2["info"]["phone_number"]:
    print(phone)
```

```python
for phone in dict2["info"]["phone_number"]:
    if phone[0] == "0":
        print(f"手機:{phone}")
    else:
        print(f"手機2:{phone}")
```

##### dict 的合併

```python
# dict 的合併
fruit_info = {
    "apple" : {"price" : 80, "quantity":100, "color":["green","red"]},
    "orange" : {"price" : 50, "quantity":120,"color":["orange"]},
    "tomato" : {"price" : 40, "quantity":150,"color":["red"]}
}

data = {
    "banana" : {"prince" : 40 , "quantity":90 , "color":["yello"]}
}

print(fruit_info)
print(data)
```

```python
# dict 的合併 
from pprint import pprint
fruit_info.update(data)
pprint(fruit_info)

# x.update(y)
# 兩dict x,y合併
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
from pprint import pprint
pprint(fruit_info)
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''

print(fruit_info)

if "yellow" not in fruit_info["apple"]["color"]:
    fruit_info["apple"]["color"].append("yellow")
if "green" not in fruit_info["tomato"]["color"]:
    fruit_info["tomato"]["color"].append("green")
# fruit_info["tomato"]["color"].append("green")
pprint(fruit_info)
# 會重複執行 會有好幾個yellow
# 補救辦法如下：
while fruit_info["apple"]["color"].count("yellow") > 1:
    fruit_info["apple"]["color"].remove("yellow")

pprint(fruit_info)
```
